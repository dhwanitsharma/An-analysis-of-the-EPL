```python
df_GCA = pd.read_csv('Goal and Shot Creation.csv')
df_GCA
```

```python
df_GCA1 = df_GCA[df_GCA['Season'] != '2020-2021']
df_GCA2 = df_GCA1[df_GCA1['Player'] == 'Squad Total']
df_GCA3 = df_GCA2[df_GCA2.Team.isin (['Manchester United','Leicester City','Burnley','Southampton','Everton','West Ham United','Watford','Bournemouth'])]  
df_GCA3 = df_GCA3.drop(columns = ['Nation','Pos','Age','90s','Player'])
df_GCA3
```

```python
df_GCA3_normed = df_GCA3.copy()
df_GCA3_normed
for col in df_GCA3_normed.columns[1:-2]:
    df_GCA3_normed[col]  = df_GCA3_normed[col].astype(float)
    df_GCA3_normed[col] = (df_GCA3_normed[col]-df_GCA3_normed[col].min())/ (df_GCA3_normed[col].max()-df_GCA3_normed[col].min())

vars_GCA3 = df_GCA3_normed[df_GCA3_normed.columns[1:-2]].var().sort_values(ascending = False)
vars_GCA3
```

```python
important_stats = vars_GCA3.index[:4]
important_stats
slice = df_GCA3[list(important_stats)+['Team', 'Season']]
slice.head()
width = .8
for col in slice.columns[:4]:
    plt.figure(figsize=(12, 8), dpi=60)
    plt.bar(slice[slice['Season']=='2018-2019']['Team'],slice[slice['Season']=='2018-2019'][col] , width=width, 
        color='b', label='2018-2019')
    plt.bar( slice[slice['Season']=='2019-2020']['Team'], slice[slice['Season']=='2018-2019'][col], 
        width=0.5*width, color='r', alpha=0.5, label='2019-2020')
    plt.xticks(Rotation = 90)
    plt.title(col)
    plt.legend(loc="upper left")
```
